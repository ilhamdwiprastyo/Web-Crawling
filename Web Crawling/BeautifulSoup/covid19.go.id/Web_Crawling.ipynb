```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://covid19.go.id/p/berita?page='
count =0
hasil=[]
for x in range(1,57):
    url_new = url+str(x)
    

    # Membuat requests
    r = requests.get(url_new)

    # Hasil response
    request = r.content

    soup = BeautifulSoup(request, 'html.parser')

    # Extract Element
    title = soup.findAll('a', attrs={'class':'text-color-dark'})

    for x in range(0, len(title)):
        count += 1
        hasil.append(title[x].text.strip())
        print("{0}. {1}\n".format(count, title[x].text.strip()))
        
array = np.array(hasil)
df = pd.DataFrame(array, columns=['Judul'])    
```
